# Example `h5x` Construction of Custom ATL09 DataFrame

```python
from sliderule import sliderule, h5
```

```python
session = sliderule.create_session(domain="localhost", cluster=None, verbose=True)
```

```python
asset = "icesat2-atl09"
resource = "ATL09_20250302214125_11732601_006_01.h5"
variables = [
    "bsnow_h",
    "cap_h",
    "delta_time",
    "latitude",
    "longitude"
]
groups = [
    "/profile_1/low_rate",
    "/profile_2/low_rate",
    "/profile_3/low_rate"
]
gdf = h5.h5x(variables, resource, asset, groups, time_column="delta_time", x_column="longitude", y_column="latitude", session=session)
```
